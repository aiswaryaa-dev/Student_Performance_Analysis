```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Aiswaryaa\OneDrive\Documents\Student_Data.csv")
print(df)
```

<#check dataset information#>

```python
print(df.info())
```

<#checking missing values#>

```python
print(df.isnull().sum())
```

<#calculate total marks#>

```python
df["Total"] = df["Tamil"] + df["English"] + df["Maths"] + df["Science"] + df["Social"]
print(df)
```

<#calculate average marks#>

```python
df["Average"] = df["Total"] / 3
print(df)
```

<#find top scorer#>

```python
top_student =df.loc[df["Total"].idxmax()]
print("Top Scorer:")
print(top_student)
```

<#find lowest scorer#>

```python
lowest_student = df.loc[df["Total"].idxmin()]
print("Lowest Scorer:")
print(lowest_student)
```

<#find subject wise average#>

```python
print("Tamil average:",df["Tamil"].mean())
print("English average:",df["English"].mean())
print("Maths average:",df["Maths"].mean())
print("Science average:",df["Science"].mean())
print("Social average:",df["Social"].mean())
```

<#pass/fail analysis#>

```python
df["Result"] = df["Total"].apply(lambda x:"pass" if x >=120 else"Fail")
print(df[["Name","Total","Result"]])
```

<#count pass and fail students#>

```python
print(df["Result"].value_counts())
```

<#gender-wise average#>

```python
print(df.groupby("Gender")["Average"].mean())
```

<#create bar chart#>

```python
plt.figure(figsize=(8,5))
plt.bar(df["Name"],df["Total"])
plt.title("Student Total Marks")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.xticks(rotation=45)
plt.show()
```

<# create pie chart #>

```python
result_count = df["Result"].value_counts()
plt.pie(result_count,labels=result_count.index,autopct="%1.1f%%")
plt.title("pass and fail percentage")
plt.show()
```

<#final report#>

```python
print("----STUDENT PERFORMANCE REPORT----")
print("\nTop scorer:")
print(df.loc[df["Total"].idxmax()])
print("\nLowest scorer:")
print(df.loc[df["Total"].idxmax()])
print("\nClass average:")
print(df["Average"].mean())

```
